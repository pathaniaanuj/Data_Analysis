```python
# need top 10 countries having most confirmed , death and recovering cases
confirm_death_data=covid.loc[(covid['Confirmed']>40000 )&(covid['Deaths']>2000)& (covid["Recovered"]>30000)]
```

```python
confirm_death_data
```

```python
confirm_death_data=pd.DataFrame(confirm_death_data)
```

```python
confirm_death_data.shape # in this data we have 24 rows with our specific condition
```

```python
#sort tha data on having highest confirmed case
confirm_death_data.sort_values(by=['Confirmed','Deaths','Recovered'],ascending=False,inplace=True)
```

```python
top_10_countries=confirm_death_data.head(10)
```

```python
top_10_countries
```

```python
x=top_10_countries['Confirmed']
```

```python
x
```

```python
y=top_10_countries["Deaths"]
```

```python
z=top_10_countries["Recovered"]
```

```python
pip install pandas matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,top_10_countries["Deaths"],marker='o',color='green', linestyle='-', linewidth=3)
plt.xlabel('Deaths')
plt.ylabel('Confirmed cases')
plt.title('covid cases with deaths')
plt.grid(True)
plt.fill_between(x, y, where=(x > y), interpolate=True, color='gray', alpha=0.3)
plt.show()
```

```python
plt.scatter(x,z,marker='o',color='blue', linestyle='-', linewidth=3)
plt.xlabel('Recoverd cases')
plt.ylabel('Confirmed cases')
plt.title('covid cases with recovering')
plt.grid(True)
plt.show()
```
